# Contrast Enhancement

```python
import cv2
import matplotlib.pyplot as plt
```

```python
def enhance_contrast(image_path):
  # Read the image in grayscale
  img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
  if img is None:
    print("Error: Could not read image.")
    return

  # Apply histogram equalization
  enhanced_img = cv2.equalizeHist(img)

  # Display original and enhanced images
  plt.figure(figsize=(10, 5))
  plt.subplot(1, 2, 1)
  plt.imshow(img, cmap='gray')
  plt.title('Original Image')
  plt.axis('off')

  plt.subplot(1, 2, 2)
  plt.imshow(enhanced_img, cmap='gray')
  plt.title('Enhanced Image')
  plt.axis('off')
  plt.show()

# Example usage
enhance_contrast('input.jpg')
```

# Histogram Equalization

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def histogram_equalization(image_path):
  # Read the image in grayscale
  img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

  if img is None:
    print("Error: Could not read image.")
    return

  # Apply histogram equalization
  enhanced_img = cv2.equalizeHist(img)

  # Compute histograms
  original_hist = cv2.calcHist([img], [0], None, [256], [0, 256])
  enhanced_hist = cv2.calcHist([enhanced_img], [0], None, [256], [0, 256])

  # Display images and histograms
  plt.figure(figsize=(12, 6))

  plt.subplot(2, 2, 1)
  plt.imshow(img, cmap='gray')
  plt.title('Original Image')
  plt.axis('off')

  plt.subplot(2, 2, 2)
  plt.imshow(enhanced_img, cmap='gray')
  plt.title('Enhanced Image')
  plt.axis('off')

  plt.subplot(2, 2, 3)
  plt.plot(original_hist, color='black')
  plt.title('Original Histogram')

  plt.subplot(2, 2, 4)
  plt.plot(enhanced_hist, color='black')
  plt.title('Enhanced Histogram')

  plt.show()

# Example usage
histogram_equalization('input.jpg')
```

# Adaptive Filter

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import wiener
```

```python
def adaptive_wiener_filter(image_path):
    # Read the image in grayscale
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    if img is None:
        print("Error: Could not read image.")
        return

    # Convert to float32 without normalizing
    img_float = img.astype(np.float32)

    # Apply Wiener filter
    filtered_img = wiener(img_float, (5, 5))

    # Clip and convert back to uint8
    filtered_img = np.clip(filtered_img, 0, 255).astype(np.uint8)

    # Display original and filtered images
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(img, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(filtered_img, cmap='gray')
    plt.title('Wiener Filtered Image')
    plt.axis('off')

    plt.show()

# Example usage
adaptive_wiener_filter('input.jpg')
```

### What do you think is the importance of a histogram when it comes to image processing? Cite real life usage of histogram and explain it comprehensively.
Histograms display the frequency of each intensity value in an image, making them essential for analyzing brightness, contrast, and tonal distribution. They help identify areas for improvement, enhance image quality, and detect objects. One of their most significant applications is in medical imaging, where X-ray and MRI scans rely on contrast adjustments to improve the visibility of organs, tissues, and potential abnormalities. For instance, in radiology, histograms assist in highlighting differences in tissue density, making it easier to detect fractures, tumors, or infections.

### On Task 3, what do you think is the main difference of Wiener filter, compared to the earlier filters you have used?
The Wiener filter is commonly used for noise reduction and image deblurring, relying on statistical properties of the signal and noise. In other words, it restores the quality of images degraded by noise or blur. Unlike other filtering techniques, which may overly smooth the image or introduce artifacts, the Wiener filter preserves important details while minimizing distortion. This is why its output often appears most similar to the original image.
